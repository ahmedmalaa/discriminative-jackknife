```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 

from models.static import *

from influence.influence_computation import *
from influence.influence_utils import *


def compare_IF_LOO(model, IF_compute_mode="exact", damp=0):
    
    IF_losses     = []  
    exact_losses  = []
    n_data_points = model.X.shape[0]
    
    for k in range(n_data_points):
        
        print(k)
        
        # compute influence functions
        
        if IF_compute_mode=="exact":
            
            IF   = influence_function(model, train_index=[k], mode="exact", damp=damp)[0]
            
            #H_ij = exact_hessian_ij(model, model.loss_fn(model.predict(model.X[k, :], numpy_output=False), model.y[k]))
            #H    = exact_hessian(model)
            #IF2  = torch.mm(torch.inverse(H), torch.mm(H_ij, IF)) * (2 / model.X.shape[0])
        
        else:
            
            IF = influence_function(model, 
                        train_index=[k], 
                        mode="stochastic", 
                        batch_size=50, 
                        damp=damp, 
                        scale=100, 
                        recursion_depth=10000)[0]
        
        # perturb the original model using IFs 
        
        model_perturbed = perturb_model_(model, IF)
        model_perturbed.update_loss()
        
        IF_losses.append(model_perturbed.loss.detach().numpy())
        
        # re-train the model with one data point deleted
        
        model_perturbed_exact = DNN(num_layers=1, num_hidden=20, n_dim=n_dim , activation=activtn)

        model_perturbed_exact.fit(np.delete(x_train, obj=k, axis=0), 
                                  np.delete(y_train, obj=k, axis=0), 
                                  learning_rate=0.01, num_iter=n_epochs, verbosity=False)
        
        # evaluate exact LOO loss
        exact_loss = model.loss_fn(model_perturbed_exact.predict(model.X, numpy_output=False), model.y)
        
        exact_losses.append(exact_loss.detach().numpy())
        
    return IF_losses, exact_losses
```

```python
def linear_data_model(x, beta, sigma):
    
    y = np.sum(x * np.repeat(beta.reshape((-1, 1)), repeats=x.shape[0], axis=1).T, axis=1).reshape(-1, 1) + (sigma**2) * np.random.normal(size=x.shape[0]).reshape(-1, 1)
    
    return y
```

```python
# create dummy data for training
n_sample = 100
n_epochs = 500
n_dim    = 1
sigma    = 5
beta     = np.array([3]) #np.array([3, 2, 1, -1, -2])
activtn  = "Tanh"
damp     = 0

x_train  = np.random.normal(0, 1, (n_sample, n_dim))
y_train  = linear_data_model(x_train, beta, sigma)
```

```python
model    = DNN(num_layers=1, num_hidden=20, n_dim=n_dim , activation=activtn)

model.fit(x_train, y_train, learning_rate=0.01, num_iter=n_epochs, verbosity=True)
```

```python
params_ = []

for param in model.parameters():
    
    print(param)
    
    params_.append(param)
```

```python
#losses correlation
```

```python
L_IF, L_exact = compare_IF_LOO(model, damp=damp)
```

```python
loss_range    = np.linspace(np.minimum(np.min(np.array(L_IF)), np.min(np.array(L_exact))) - 1, 
                            np.maximum(np.max(np.array(L_IF)), np.max(np.array(L_exact))) + 1, 100)

plt.scatter(L_IF, L_exact, color="red")
plt.plot(loss_range, loss_range, color="black", linestyle=":")

plt.xlim(520, 570)
plt.ylim(520, 570)
```

```python
data_index = 6
IF         = influence_function(model, train_index=[data_index], mode="exact", damp=damp)
```

```python
model_perturbed       = perturb_model_(model, IF[0])
model_perturbed.update_loss()

# re-train the model with one data point deleted
        
model_perturbed_exact = DNN(num_layers=1, num_hidden=20, n_dim=n_dim , activation="Tanh")

model_perturbed_exact.fit(np.delete(x_train, obj=data_index, axis=0), 
                          np.delete(y_train, obj=data_index, axis=0), 
                          learning_rate=0.01, num_iter=n_epochs, verbosity=False)
        
```

```python
x_test          = np.array(np.linspace(-5, 5, 100), dtype=np.float32)
x_test          = x_test.reshape(-1, 1)

y_pred          = model.predict(x_test)
y_pred_perturb  = model_perturbed.predict(x_test)
y_exact_perturb = model_perturbed_exact.predict(x_test)

plt.clf()
plt.plot(x_train, y_train, 'go', label='True data', alpha=0.75)
plt.plot(x_test, y_pred, '--', label='Predictions', linewidth=2, color="g", alpha=0.85)
plt.plot(x_test, y_pred_perturb, ':', linewidth=2, color="r", label='Perturbed predictions', alpha=0.75)
plt.plot(x_test, y_exact_perturb, ':', linewidth=2, color="b", label='Exact LOO predictions', alpha=0.75)
plt.plot(x_train[data_index], y_train[data_index], 'go', color="r", label='Deleted data')
plt.legend(loc='best')
plt.show()
```
